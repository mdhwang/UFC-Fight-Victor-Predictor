# Winner Stats vs Loser Stats

```python
blue_cols = [
 'Winner',
 'weight_class',
 'B_age',
 'B_Height_cms',
 'B_Reach_cms_Updates',
 'B_Weight_lbs']
red_cols = [
 'Winner',
 'weight_class',
 'R_age',
 'R_Height_cms',
 'R_Reach_cms_Updates',
 'R_Weight_lbs']


reduced_red = df[red_cols]
reduced_blue = df[blue_cols]
print(len(reduced_red))
print(len(reduced_blue))
```

```python
red_win = reduced_red[reduced_red["Winner"] == "Red"]
red_lose = reduced_red[reduced_red["Winner"] != "Red"]
blue_win = reduced_blue[reduced_red["Winner"] == "Blue"]
blue_lose = reduced_blue[reduced_red["Winner"] != "Blue"]

print(len(red_win))
print(len(red_lose))
print(len(blue_win))
print(len(blue_lose))
```

```python
headers = ["Winner","Weight Class","Age","Height","Reach","Weight"]

red_win.columns = headers
blue_win.columns = headers
red_lose.columns = headers
blue_lose.columns = headers

red_lose.head()
```

```python
winners = red_win.append(blue_win)
losers = red_lose.append(blue_lose)
```

```python
#Reach Diff
winners.groupby("Weight Class")["Reach"].mean() - losers.groupby("Weight Class")["Reach"].mean()
```
